```julia
using Gadfly
using RDatasets
```

```julia
Gadfly.set_default_plot_size(800px, 600px)
```

```julia
iris = RDatasets.dataset("datasets", "iris");
```

```julia
first(iris)
```

```julia
p1 = Gadfly.plot(iris, x=:SepalLength, y=:SepalWidth, color=:Species, Geom.point)
p2 = Gadfly.plot(iris, x=:PetalLength, y=:PetalWidth, color=:Species, Geom.point)
vstack(p1, p2)
```
