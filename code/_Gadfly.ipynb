```julia
using Gadfly
using RDatasets
```

```julia
Gadfly.set_default_plot_size(800px, 600px)
```

```julia
iris = dataset("datasets", "iris");
```

```julia
# Gadfly
p1 = Gadfly.plot(iris, x=:SepalLength, y=:SepalWidth, color=:Species, Geom.point)
p2 = Gadfly.plot(iris, x=:PetalLength, y=:PetalWidth, color=:Species, Geom.point)
vstack(p1, p2)
```

```julia
Gadfly.plot(
    iris,
    x=:Species,
    y=:PetalLength,
    Geom.boxplot
)
```

```julia
salaries = dataset("car","Salaries");
```

```julia
Gadfly.plot(
    salaries,
    x=:Discipline,
    y=:Salary,
    color=:Rank,
    Geom.boxplot
)
```
